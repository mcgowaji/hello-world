```python
#imports
import numpy as np
import pandas as pd
```

```python
#Read in and display main dataframe
movies_raw= pd.read_csv('movies_metadata.csv', engine='python', encoding='utf-8')
movies_raw.head()
```

```python

ratings=pd.read_csv('ratings.csv')
ratings.head()
```

```python
ratings.info()
```

```python
#Equate indices
ratings.movieId=ratings.movieId-1
```

```python
#Create aggregates to determine average ratings and vote counts per film
rating_aggs=pd.DataFrame()
rating_aggs['average_rating']=ratings.groupby('movieId').mean()['rating']
rating_aggs['num_votes']= ratings.groupby('movieId').count()['userId']
```

```python
rating_aggs.head()
```

```python
#Read in Ratings DataFrame
links=pd.read_csv('links.csv')
links.head()
```

```python
#Join links and movies
#First we need to clean movie_ids and imdb_ids to match
movies_raw.imdb_id= movies_raw.imdb_id.str.replace('tt0', '')
links.movieId= links.movieId-1
```

```python
#Join movies0 and ratings
movies=pd.concat([movies_raw, rating_aggs], axis=1)
movies.head()
```

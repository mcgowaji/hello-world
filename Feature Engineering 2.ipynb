3.  Now, apply PCA to the 4 expenditure types. How much of the total variance is explained by the 1st component?

```python
#Remove non-floats for PCA
education= education.drop(columns=['STATE', 'PRIMARY_KEY'])
```

```python
#Normalize Data to have mean of 0, standard deviation of 1.
education_expenditures= education[['TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE', 'SUPPORT_SERVICES_EXPENDITURE', 'OTHER_EXPENDITURE']]
X = StandardScaler().fit_transform(education_expenditures)
```

```python
#Run PCA on expenditure components
sklearn_pca= PCA(n_components=1)
Y_sklearn= sklearn_pca.fit_transform(X)

print(
    'The percentage of total variance in the dataset explained by each',
    'component calculated by SKLearn.\n',
   sklearn_pca.explained_variance_ratio_
)
```

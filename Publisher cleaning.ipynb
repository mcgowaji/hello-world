```python
#Clean cost column
def cost_to_float(x):
    try:
        return float(x)
    except ValueError:
        return 2000
df['COST']= df['COST'].apply(cost_to_float)
```

```python
#Identify 5 most common publishers
most_common=df.groupby('cleanpub').count().sort_values(by='Publisher', ascending=0).reset_index()[:5]
most_common
```

```python
#Replace high outliers with median
median = df.loc[df['COST']< 9000, 'COST'].median()
df.loc[df.COST > 9000, 'COST'] = np.nan
df.fillna(median,inplace=True)
```

```python
#Calculate mean, median, std for each of the top 5 publishers
stats_df= df.groupby('cleanpub').agg({'COST': ['count', np.mean, np.median, np.std]}).sort_values(by=('COST', 'count'), ascending=0)[:5] 
stats_df
```
